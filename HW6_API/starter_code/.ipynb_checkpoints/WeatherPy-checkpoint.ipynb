```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

import requests
import json
from api_keys import api_key
from citipy import citipy
```

# Randomly select 500 unique cities based on latitude and longitude

```python
# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

#Api Key for Open Weather
key = api_key
```

```python
cities = []
for i in range (2000):
    random_latitude = np.random.uniform(lat_range[0],lat_range[1])
    random_longitude = np.random.uniform(lng_range[0],lat_range[1])
    city = citipy.nearest_city(random_latitude, random_longitude).city_name
    
    base_url = 'http://api.openweathermap.org/data/2.5/find?q='+ city + '&units=imperial&appid=' + api_key 
    res = requests.get(base_url)
    response = res.json()
    
    if not response['list']:
        continue
        
    if city not in cities:
        cities.append(city)
    
    if len(cities) == 500 :
        break 
```

# Perform a weather check on each of the cities

* Temperature (F) vs. Latitude
* Humidity (%) vs. Latitude
* Cloudiness (%) vs. Latitude
* Wind Speed (mph) vs. Latitude

```python
data = pd.DataFrame(columns = ['city', 'lat', 'long', 'temp(F)', 'humidity(%)', 'cloudiness(%)', 'wind_speed'])
data
```

```python
city = []
latitude_list = []
longititude_list = []
tempreture_list = []
humidity_list = []
windspeed_list = []
cloudinesss_list = []
```

```python
for i in cities:
    
    base_url = 'http://api.openweathermap.org/data/2.5/find?q='+ i + '&units=imperial&appid=' + api_key 
    res = requests.get(base_url)
    response = res.json()
        
        
    city.append(i)

    city_lon = response['list'][0]['coord']['lon']
    longititude_list.append(city_lon)

    city_lat = response['list'][0]['coord']['lat']
    latitude_list.append(city_lat)

    city_temp = response['list'][0]['main']['temp']
    tempreture_list.append(city_temp)

    city_humidity = response['list'][0]['main']['humidity']
    humidity_list.append(city_humidity)

    city_windspeed = response['list'][0]['wind']['speed']
    windspeed_list.append(city_windspeed)

    city_cloudiness = response['list'][0]['clouds']['all']
    cloudinesss_list.append(city_cloudiness)

  
    
```

```python
data['city'] = city
data['lat'] = latitude_list
data['long'] = longititude_list
data['temp(F)'] = tempreture_list
data['humidity(%)'] = humidity_list
data['cloudiness(%)'] = cloudinesss_list
data['wind_speed'] = windspeed_list
```
